# Basic Visualization Rules

1. The first step is to choose the appropriate plot type. 
2. Second, when we choose type of plot, one of the most important things is to label your axis. If we donâ€™t do this, the plot is not informative enough. 
3. We can add a title to make our plot more informative.
4. Add labels for different categories when needed.
5. Annotation, optionally we can add a text or an arrow at interesting data points.
6. In some cases we can use some sizes and colors of the data to make the plot more informative.

# What are type of plots in matplotlib in python?
* Line plot - very useful when it comes to multivariate analysis.
* Histograms - to understand the distribution.
* Bar charts - we can use to compare one over the other.
* Pie charts - to present the data in percentage, ratio, proportion etc.
* Scatterplots - to identify the relationship between two variables.
* Subplots - Very useful when comparing two or more plots

# Pie Chart

* pie chart is used for to present the data in percentage, ratio, proportion etc.

```python
#This is how we can create a pie chart
location = ['Maharastra', 'Gujarat', 'Tamilnadu', 'Delhi', 'Rajasthan', 'Madhya Pradesh', 'Uttar pradesh','odisha']
Confirmed_cases = [23000,8500, 8000,7200,4000, 3700, 3500,2500]
plt.pie(Confirmed_cases,labels=location,labeldistance= 1.1,radius= 1.5,shadow = True,counterclock= True)
```

```python
plt.figure(figsize=(5,6))
#For style
plt.style.use('tableau-colorblind10')
plt.axis('equal')
plt.pie(Confirmed_cases, labels= location, explode=[0,0.1,0,0,0,0,0,0],radius= 1.5, autopct= '%2.1f%%', pctdistance = 0.85,
        shadow = True, labeldistance= 1.1, counterclock= True, wedgeprops={'width': 0.5})
plt.title("Positive cases")
plt.show()
```

# Histogram

* It used for to understand the distribution.

```python
Confirmed_cases = [23000,8500, 8000,7200,4000, 3700, 3500,2500,4837,3783,8209,7264,7372]
plt.hist(Confirmed_cases,range=[1000,5000], density=False, bins= 6, rwidth=0.85,
         histtype='barstacked', label= 'day_wise_cases', color= 'grey')
plt.title('Day-wise-count')
plt.xlabel('Number of cases')
plt.ylabel('counts')
plt.legend()
```

# Scatter plot

* it used to identify the relationship between two variables.

```python
plt.style.use('ggplot')
plt.figure(figsize=(10,8))
day_wise_cases = [1755,2411,2487,2573,3875,2680, 3561,3390,3320,3277,4213]
Deaths = [80, 113, 125, 130,258,146,270, 230,228,243,330 ]

plt.scatter(day_wise_cases, Deaths, label = 'Scatter plot', marker = '+', s= 100, linewidths=2.0, alpha= 0.7)
#title of the plot
plt.title("Positive-cases Vs Deaths",fontdict={'size':20, 'color':'black', 'weight':'black'}, fontsize = 15, color = 'black')
#set x-label& y-label
plt.xlabel('day-wise-positive-cases',fontdict={'size':20, 'color':'black', 'weight':'black'}, fontsize= 15, color = 'black')
plt.ylabel('Day-wise-death-count', fontdict={'size':20, 'color':'black', 'weight':'black'},fontsize = 15, color = 'black')

plt.legend()
plt.show()
```
